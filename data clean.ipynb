```python
import pandas as pd

df_class = pd.read_csv("Student_Placement_Skills_2025.csv")

# 先简单检查数据的基本情况，检查数据有多少行多少列，有没有空值和重复值
print("原始数据大小：", df_class.shape)
print("\n每一列的空值数量：")
print(df_class.isnull().sum())
print("\n重复行数量：", df_class.duplicated().sum())

# 统一文字格式
df_class["Gender"] = df_class["Gender"].astype(str).str.strip().str.title()
df_class["Placement_Offer"] = df_class["Placement_Offer"].astype(str).str.strip().str.title()
df_class["Degree"] = df_class["Degree"].astype(str).str.strip()

# 删除 Student_ID，学生编号，对预测没有实际帮助
df_class = df_class.drop(columns=["Student_ID"])

# 把 Gender 变成数字， Male 记为 1，Female 记为 0
df_class["Gender"] = df_class["Gender"].map({"Male": 1, "Female": 0})

#把 Placement_Offer 也变成数字，Yes 记为 1，No 记为 0
df_class["Placement_Offer"] = df_class["Placement_Offer"].map({"Yes": 1, "No": 0})

# 删除 Salary_Offered_USD，这个任务是预测有没有拿到 offer，如果把薪资也放进去，就相当于提前告诉模型结果，不合理
df_class = df_class.drop(columns=["Salary_Offered_USD"])

# 把 Degree 做成 one-hot encoding，因为专业是文字，不能直接给模型使用
df_class = pd.get_dummies(df_class, columns=["Degree"], dtype=int)

# 检查一下处理后的数据
print("\n处理后数据大小：", df_class.shape)
print("\n处理后每一列的空值数量：")
print(df_class.isnull().sum())

# 看前五行，确认数据有没有处理成功
print("\n处理后的前五行数据：")
display(df_class.head())

# 保存成新的 csv 文件
df_class.to_csv("classification_encoded.csv", index=False)

print("\n分类任务的数据已经保存好了：classification_encoded.csv")
```

```python
import pandas as pd

df_reg = pd.read_csv("Student_Placement_Skills_2025.csv")

# 检查原始数据
print("原始数据大小：", df_reg.shape)
print("\n每一列的空值数量：")
print(df_reg.isnull().sum())
print("\n重复行数量：", df_reg.duplicated().sum())

# 统一文字格式
df_reg["Gender"] = df_reg["Gender"].astype(str).str.strip().str.title()
df_reg["Placement_Offer"] = df_reg["Placement_Offer"].astype(str).str.strip().str.title()
df_reg["Degree"] = df_reg["Degree"].astype(str).str.strip()

# 删除 Student_ID
df_reg = df_reg.drop(columns=["Student_ID"])

# 只保留拿到 offer 的学生,薪资预测只对已经拿到 offer 的学生有意义
df_reg = df_reg[df_reg["Placement_Offer"] == "Yes"].copy()

# 把 Gender 变成数字,Male 记为 1，Female 记为 0
df_reg["Gender"] = df_reg["Gender"].map({"Male": 1, "Female": 0})

# 第七步：删除 Placement_Offer,现在留下来的都是 Yes，这一列已经没有作用了
df_reg = df_reg.drop(columns=["Placement_Offer"])

# 把 Degree 做成 one-hot encoding,把文字专业拆成多个 0/1 的列，方便后面建模
df_reg = pd.get_dummies(df_reg, columns=["Degree"], dtype=int)

# 检查处理后的结果
print("\n处理后数据大小：", df_reg.shape)
print("\n处理后每一列的空值数量：")
print(df_reg.isnull().sum())

# 看前五行，确认数据处理有没有成功
print("\n处理后的前五行数据：")
display(df_reg.head())

# 保存成新的 csv 文件
df_reg.to_csv("regression_encoded.csv", index=False)

print("\n回归任务的数据已经保存好了：regression_encoded.csv")
```
