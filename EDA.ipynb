```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("data clean for EDA.csv")
sns.set(style="whitegrid")
```

```
df.columns
```

```
df.info()
```

```
df.describe()
```

```
offer_counts = df['Placement_Offer'].value_counts()
offer_pct = df['Placement_Offer'].value_counts(normalize=True) * 100
print(f"Yes: {offer_counts['Yes']} 人（{offer_pct['Yes']:.1f}%）")
print(f"No:  {offer_counts['No']} 人（{offer_pct['No']:.1f}%）")

sns.countplot(x='Placement_Offer', data=df)
plt.title("Placement Offer Distribution")
plt.xlabel("Placement Offer")
plt.ylabel("Count")
plt.show()
```

```
salary = df['Salary_Offered_USD']
print(f"均值:   ${salary.mean():,.1f}")
print(f"中位数: ${salary.median():,.1f}")
print(f"标准差: ${salary.std():,.1f}")
print(f"最小值: ${salary.min():,.1f}")
print(f"最大值: ${salary.max():,.1f}")
print(f"偏度:   {salary.skew():.3f}")

sns.histplot(df['Salary_Offered_USD'], kde=True)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```
cgpa = df['CGPA']
print(f"均值:   {cgpa.mean():.3f}")
print(f"中位数: {cgpa.median():.3f}")
print(f"标准差: {cgpa.std():.3f}")
print(f"最小值: {cgpa.min():.2f}")
print(f"最大值: {cgpa.max():.2f}")
print(f"偏度:   {cgpa.skew():.3f}")

sns.histplot(df['CGPA'], kde=True)
plt.title("CGPA Distribution")
plt.xlabel("CGPA")
plt.ylabel("Frequency")
plt.show()
```

```
degree_counts = df['Degree'].value_counts()
for deg, count in degree_counts.items():
    print(f"{deg}: {count} 人")

sns.countplot(x='Degree', data=df)
plt.xticks(rotation=45)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Count")
plt.show()
```

```
yes = df[df['Placement_Offer'] == 'Yes']
no  = df[df['Placement_Offer'] == 'No']
print(f"有offer — 均值: {yes['CGPA'].mean():.2f}, 中位数: {yes['CGPA'].median():.2f}, std: {yes['CGPA'].std():.2f}")
print(f"无offer — 均值: {no['CGPA'].mean():.2f}, 中位数: {no['CGPA'].median():.2f}, std: {no['CGPA'].std():.2f}")

sns.boxplot(x='Placement_Offer', y='CGPA', data=df)
plt.title("CGPA vs Placement Offer")
plt.xlabel("Placement Offer")
plt.ylabel("CGPA")
plt.show()
```

```
col = 'Technical_Skills_Score_100'
print(f"有offer — 均值: {yes[col].mean():.2f}, 中位数: {yes[col].median():.2f}")
print(f"无offer — 均值: {no[col].mean():.2f}, 中位数: {no[col].median():.2f}")

sns.boxplot(x='Placement_Offer', y=col, data=df)
plt.title("Technical Skills vs Placement")
plt.show()
```

```
col = 'Internships_Count'
print(f"有offer — 均值: {yes[col].mean():.2f}, 中位数: {yes[col].median():.2f}")
print(f"无offer — 均值: {no[col].mean():.2f}, 中位数: {no[col].median():.2f}")

sns.boxplot(x='Placement_Offer', y=col, data=df)
plt.title("Internships vs Placement")
plt.show()
```

```
r = df['CGPA'].corr(df['Salary_Offered_USD'])
print(f"相关系数 r = {r:.4f}")

sns.regplot(x='CGPA', y='Salary_Offered_USD', data=df)
plt.title("CGPA vs Salary")
plt.show()
```

```
r = df['Technical_Skills_Score_100'].corr(df['Salary_Offered_USD'])
print(f"相关系数 r = {r:.4f}")

sns.regplot(x='Technical_Skills_Score_100', y='Salary_Offered_USD', data=df)
plt.title("Skills vs Salary")
plt.show()
```

```

for deg, grp in df.groupby('Degree'):
    rate = (grp['Placement_Offer'] == 'Yes').mean() * 100
    n_yes = (grp['Placement_Offer'] == 'Yes').sum()
    print(f"{deg}: {rate:.1f}% ({n_yes}/{len(grp)})")

sns.countplot(x='Degree', hue='Placement_Offer', data=df)
plt.xticks(rotation=45)
plt.title("Placement by Degree")
plt.show()
```

```

numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()

print("各变量与薪资的相关系数：")
print(corr['Salary_Offered_USD'].sort_values(ascending=False).round(4))
```

```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
